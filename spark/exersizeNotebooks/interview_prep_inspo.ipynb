```python
# 1. Data Cleaning

from pyspark.sql import SparkSession, functions as func
from pyspark.sql.types import StructField, StringType, StructType, IntegerType


spark = SparkSession.builder.appName("cleaning time").getOrCreate()

schema = StructType([
    StructField("id", IntegerType()),
    StructField("name", StringType()),
    StructField("age", IntegerType()),
    StructField("city", StringType()),
])


df = spark.read.schema(schema).option("header", True).csv("../workdayData/peeps.csv")
df = df.drop_duplicates()
avg = int(df.agg(func.round(func.avg(func.col("age")))).first()[0])
df = df.na.fill(avg, subset=["age"])
df = df.filter(func.col("age") > 30)
df.show()
spark.stop()
```

```python
# Problem #6
from pyspark.sql import SparkSession, functions as functions
from pyspark.sql.types import StructField, StructType, StringType, IntegerType, DateType
from pyspark.sql.window import Window

spark = SparkSession.builder.appName('tugnjekmjejkbn dj===').getOrCreate()
schema = StructType([
    StructField("user", StringType()),
    StructField("date", DateType()),
    StructField("amount", IntegerType())
])
df = spark.read.schema(schema).option("header", True).csv("../workdayData/orders_by_day.csv")

window = Window.partitionBy("user").orderBy("date")
window2 = Window.partitionBy("user").orderBy("date").rowsBetween(-6,0)


running_total = df.withColumn(
    "running_total",
    func.sum("amount").over(window)
)

running_total.show()

previous = df.withColumn(
    "previous",
    func.lag("amount").over(window)
)

previous.show()

difference = df.withColumn(
    "difference",
    func.col("amount") - func.lag("amount",1,0).over(window)
)

difference.show()


seven_day_avg = df.withColumn(
    "running_total",
    func.round(
        func.avg("amount").over(window2),
        2
    )

)

seven_day_avg.show()

spark.stop()
```

```python
# problem #7
from pyspark.sql import SparkSession, functions as functions
from pyspark.sql.types import StructField, StructType, StringType, TimestampType
from pyspark.sql.window import Window

spark = SparkSession.builder.appName('tugnjekmjejkbn dj===').getOrCreate()
schema = StructType([
    StructField("user", StringType()),
    StructField("timestamp", TimestampType())
])
df = spark.read.schema(schema).option("header", True).csv("../workdayData/timestamps.csv")

window = Window.partitionBy("user").orderBy("timestamp")

sessions = df.withColumn(
    "prev_timestamp",
    func.lag("timestamp").over(window)
).withColumn(
    "session_start",
    func.when(
        (func.col("prev_timestamp").isNull()) |
        ((func.unix_timestamp("timestamp") - func.unix_timestamp("prev_timestamp")) > 1800),
        1
    ).otherwise(0)
)

window = Window.partitionBy("user").orderBy("timestamp")


sessions = sessions.withColumn(
    "session_num",
    func.sum("session_start").over(window)
)

sessions = sessions.select("user", "timestamp", "session_num")
sessions.show()

spark.stop()

```

```python
# Problem 15 (THE BIG BOY)


from pyspark.sql import SparkSession, functions as func
from pyspark.sql.types import *
from pyspark.sql.window import Window


spark = SparkSession.builder.appName("The Big Boi").getOrCreate()
spark.sparkContext.setLogLevel("WARN")

custy_schema = StructType([
    StructField("custy_id", IntegerType()),
    StructField("city", StringType()),
    StructField("segment", StringType())
])

order_schema = StructType([
    StructField("order_id", IntegerType()),
    StructField("custy_id", IntegerType()),
    StructField("date", DateType()),
    StructField("amount", FloatType())
])


custys = spark.read.option("header", True).schema(custy_schema).csv("../workdayData/custys15.csv")
orders = spark.read.option("header", True).schema(order_schema).csv("../workdayData/orders15.csv")

custys = custys.drop_duplicates()
orders = orders.na.drop().drop_duplicates()

joined = custys.join(orders, "custy_id", "inner")

# custys.show(61)
# orders.show(61)
# joined.show(61)

totals_by_city = joined.groupBy("city").agg(func.round(func.sum("amount"), 2).alias("total_spendings"))
totals_by_city.show()

avg_by_segment = joined.groupBy("segment").agg(func.round(func.avg("amount"), 2).alias("avg_spending"))

avg_by_segment.show()

totals_by_custy = joined.groupBy("custy_id").agg(func.round(func.sum("amount"), 2).alias("total_spendings")).orderBy("total_spendings", ascending=False)
totals_by_custy.limit(3).show()

joined.write.mode("overwrite").partitionBy("city").parquet("output/")

spark.stop()
```
