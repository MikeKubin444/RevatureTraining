```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
df = spark.read.option('inferschema', True).option('header', True).csv("../data/fakefriendsheader.csv")

df.printSchema()
print(df.show())
print(df.select(df.name, df.age).show())

spark.stop()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("customerTotals").getOrCreate()

df = spark.read.csv(
    "../data/customer-orders.csv",
    schema="custID STRING, orderID STRING, order_price DOUBLE"
)

df.createOrReplaceTempView("orders")

customer_total = spark.sql("""
    SELECT custID, SUM(order_price) AS total
    FROM orders
    GROUP BY custID
    ORDER BY total DESC          
    """)

print(customer_total.show())

spark.stop()
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as f

spark = SparkSession.builder.appName("Working With Friends").getOrCreate()
df.printSchema()
df = spark.read.option("header", True).option("inferSchema", True).csv("../data/fakefriends-header.csv")


df = df.select("age", "friends")
df = df.groupBy("age").agg(f.sum("friends").alias("total_friends"))
print(df.show())

spark.stop()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("More Dinking Around").getOrCreate()

peps = spark.read.option('header', True).option('inferSchema', True).csv("../data/fakefriends-header.csv")

print("Heres The Schema")
peps.printSchema()

peps.select('name').show()
peps.filter(peps.age <= 35).show()
peps.groupBy('age').count().show()
peps.select(peps.name, peps.age + 100).show()

spark.stop()
```
