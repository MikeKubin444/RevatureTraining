```python
from pyspark.sql import SparkSession, functions as func
 

spark = SparkSession.builder.appName("app").getOrCreate()

df = spark.read.text("../data/access_log.txt", lineSep="\n")

df2 = df.withColumn("parts", func.split("value", " "))

df3 = (
    df2.withColumn("field0", func.col("parts")[0])
       .withColumn("field1", func.col("parts")[1])
       .withColumn("field2", func.col("parts")[2])
        .withColumn("field3", func.col("parts")[3])
       .withColumn("field4", func.col("parts")[4])
       .withColumn("field5", func.col("parts")[5])
       .withColumn("field6", func.col("parts")[6])
        .withColumn("field7", func.col("parts")[7])
       .withColumn("field8", func.col("parts")[8])
       .withColumn("field9", func.col("parts")[9])
)
df3 = df3.drop("value")
df3 = df3.drop("parts")
# df3.show()
df3.show(truncate=False)
ips = df3.groupBy("field0").agg(func.count("field0").alias("count_by_ipADDR")).select("field0", "count_by_ipADDR").show()
endpoint = df3.groupBy("field6").agg(func.count("field6").alias("count_by_endpoint")).select("field6", "count_by_endpoint").show()
status = df3.groupBy("field5").agg(func.count("field5").alias("count_by_status")).select("field5", "count_by_status").show()


```

```python
from pyspark.sql import SparkSession, functions as func
from pyspark.sql.types import StructField, StructType, StringType, IntegerType
spark = SparkSession.builder.appName("app").getOrCreate()
# user id | item id | rating | timestamp

schema = StructType([
    StructField("user_id", StringType()),
    StructField("item_id", StringType()),
    StructField("rating", IntegerType()),
    StructField("timestamp", IntegerType())
])


df = spark.read.option(schema=schema).option("sep", "\t").csv("../ml-100k/u.data")
df.printSchema

df.show()
```
